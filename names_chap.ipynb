```python
from lxml import etree
from io import StringIO
import sys
import fnmatch
import os
```

```python
argsdir='/home/odysseus/Bureau/chapitres/git_repo/tagged_corp/'
files_list=fnmatch.filter(os.listdir(argsdir), '*.xml')
```

```python
with open('names.csv', 'w') as f:
    for filename_xml in files_list:

        with open(argsdir+filename_xml, 'r') as xml_file:
            xml_to_check = xml_file.read()

            print(filename_xml)

            doc = etree.parse(argsdir+filename_xml)

            index=0
            names_chap=dict()

            if doc.findall(".//div[@type='chapter']"):
                for chapter in doc.findall(".//div[@type='chapter']"):
                    names=set()
                    for word in chapter.xpath(".//word[contains(@postag,'NAME')]"):
                        if len(word.attrib['lemma'])>1 and word.attrib['postag']!='NAMEplace':
                            names.add(word.attrib['lemma'])
                    index+=1
                    names_chap[index]=names
            elif doc.findall(".//div[@type='nouvelle']"):
                for chapter in doc.findall(".//div[@type='nouvelle']"):
                    names=set()
                    for word in chapter.xpath(".//word[contains(@postag,'NAME')]"):
                        if len(word.attrib['lemma'])>1 and word.attrib['postag']!='NAMEplace':
                            names.add(word.attrib['lemma'])
                    index+=1
                    names_chap[index]=names
            elif doc.findall(".//div[@type='part']"):
                for chapter in doc.findall(".//div[@type='part']"):
                    names=set()
                    for word in chapter.xpath(".//word[contains(@postag,'NAME')]"):
                        if len(word.attrib['lemma'])>1 and word.attrib['postag']!='NAMEplace':
                            names.add(word.attrib['lemma'])
                    index+=1
                    names_chap[index]=names
            elif doc.findall(".//div[@type='book']"):
                for chapter in doc.findall(".//div[@type='book']"):
                    names=set()
                    for word in chapter.xpath(".//word[contains(@postag,'NAME')]"):
                        if len(word.attrib['lemma'])>1 and word.attrib['postag']!='NAMEplace':
                            names.add(word.attrib['lemma'])
                    index+=1
                    names_chap[index]=names
            
            f.write(filename_xml)
            for chap in names_chap.keys():
                f.write(",")
                f.write(' '.join(names_chap[chap]))
                f.write(",")
            f.write("\n")
```
